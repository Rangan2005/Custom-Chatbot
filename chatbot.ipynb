```python
from langchain.document_loaders import WebBaseLoader

# Load the webpage data
url = "https://brainlox.com/courses/category/technical"
loader = WebBaseLoader(url)
docs = loader.load()

# View extracted data
for doc in docs:
    print(doc.page_content[:500])  # Print first 500 characters of each page
```

```python
from sentence_transformers import SentenceTransformer
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load sentence-transformers model using LangChain's HuggingFaceEmbeddings wrapper
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Split the document into manageable chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = text_splitter.split_documents(docs)

# Use FAISS to store documents and their embeddings
faiss_store = FAISS.from_documents(texts, embedding_model)
faiss_store.save_local("faiss_store")
```
